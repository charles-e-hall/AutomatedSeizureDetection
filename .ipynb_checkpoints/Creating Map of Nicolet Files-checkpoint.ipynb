```python
def getNicoletFiles(path):
    allFiles = []
    allPaths = []
    nicoletFiles = []
    nicoletPaths = []

    for (dirpath, dirname, filename) in os.walk(path):
        allFiles.extend(filename)
        allPaths.extend(dirpath)
        
    for i in range(len(allFiles)):
        if allFiles[i].split('.')[-1] == 'e':
            nicoletFiles.extend(allFiles[i])
            nicoletPaths.extend(allPaths[i])
            
    del allFiles, allPaths
    return nicoletPaths, nicoletFiles
```
